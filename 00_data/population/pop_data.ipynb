```python
state_county_FIPS = pd.read_csv(
    "https://www2.census.gov/geo/docs/reference/codes2020/national_county2020.txt",
    delimiter="|",
)
state_county_FIPS.head()
```

but this is actually the same issue as earlier where alabama is 1 instead of "01", and Autauga county is 1 instead of "001"

```python
# making it a string so they are easier to concatenate
state_county_FIPS["STATEFP"] = state_county_FIPS["STATEFP"].astype(str)
state_county_FIPS["COUNTYFP"] = state_county_FIPS["COUNTYFP"].astype(str)

# adding a 0 onto the front of those that are only one digit
state_county_FIPS["STATEFP"] = state_county_FIPS["STATEFP"].astype(str).str.zfill(2)
state_county_FIPS.head()

# adding 0s as needed to make it 3 digits
state_county_FIPS["COUNTYFP"] = state_county_FIPS["COUNTYFP"].astype(str).str.zfill(3)
state_county_FIPS.sample(5)

# slayyyy
```

Okay now that I did that we need to actually concatenate to make the full FIPS code

```python
state_county_FIPS["FIP_CODE"] = (
    state_county_FIPS["STATEFP"] + state_county_FIPS["COUNTYFP"]
)
state_county_FIPS.head()
```
