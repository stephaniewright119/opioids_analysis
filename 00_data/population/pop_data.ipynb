```python
import pandas as pd
```

Data from this link https://www2.census.gov/programs-surveys/popest/datasets/2000-2010/intercensal/county/   
previously called co-est00int-tot.csv

```python
popdata = pd.read_csv("popestimate_2000_2010.csv", encoding="ISO-8859-1")
popdata.head()
```

```python
popdata_edit = popdata.copy()
popdata_edit = popdata_edit.drop(columns=["SUMLEV", "REGION", "DIVISION"])
popdata_edit.head()
```

```python
# there are rows where state and county name are the same. I am assuming this is the population of the whole state. going to save that to another data set lol
statepop_data = popdata_edit[popdata_edit["STNAME"] == popdata_edit["CTYNAME"]]
len(statepop_data)

# also going to drop the state data from the county dataset
county_popdata = popdata_edit[popdata_edit["STNAME"] != popdata_edit["CTYNAME"]]
county_popdata.head()
```
